```python
def wage(w_hours):
    return w_hours * 25

def wage_with_bonus(w_hours):
    return wage(w_hours) + 50  # 50 is the bonus
```

```python
wage(8), wage_with_bonus(8)  # (200, 250)
```

```python
def plus_five(x):
    return x + 5
    
def m_by_3(x):
    return plus_five(x) * 3

m_by_3(5) # 30
```

```python
plus_five(10), m_by_3(10)  # (15, 45)
```
