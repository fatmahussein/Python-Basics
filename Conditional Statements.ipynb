```python
def compare_to_five(y) :  # dont forget the colons and the order MATTERS
    if y < 0 :
        return "negative"
    elif y < 5 :
        return "Less"
    elif y > 5 :
        return "Greater"
    else:
        return "Equal"

print(compare_to_five(10))   # prints Greater 
```

```python
print(compare_to_five(2))    # Less
```

```python
print(compare_to_five(5))   # Equal
```

```python
print(compare_to_five(-5))   # negative
```
