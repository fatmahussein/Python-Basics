```python
# Use the print() function and the .get() method to check the price of the Spaghetti in the Price_list.
Price_list = {'Cheeseburger': 8, 'Fries': 5, 'Lasagna': 12, 'Soup': 5, 'Spaghetti': 10}

print(Price_list.get("Spaghetti")) 
 # 10
```

```python
Menu = {
    'meal_1':'Spaghetti', 
    'meal_2':'Fries', 
    'meal_3':'Cheeseburger', 
    'meal_4':'Lasagna', 
    'meal_5':'Soup'
}

prices = [10, 5, 8, 12, 5]

# Pair each menu item with its price
Price_list = dict(zip(Menu.values(), prices))
print(Price_list)   # {'Spaghetti': 10, 'Fries': 5, 'Cheeseburger': 8, 'Lasagna': 12, 'Soup': 5}
```
