```python
Food = ['Rice', 'Ugali', 'Chapati', 'Spaghetti']    # list
```

```python
Food # ['Rice', 'Ugali', 'Chapati', 'Spaghetti']
```

```python
Food[2] # 'Chapati'
```

```python
Food[3] = "Githeri"
```

```python
Food  # ['Rice', 'Ugali', 'Chapati', 'Githeri']
```

```python
del Food[3]  # not a fan of githeri so delete
```

```python
Food # ['Rice', 'Ugali', 'Chapati']
```

```python
Numbers = [10, 25, 40, 50]

print(Numbers[-3])          # 50 is at -1, 40 at -2, 30 at -3 so on and so forth
# 25
```

```python
Food.append("Pancake")   # add to the list NB: we use parenthesis here cause .append is a method
```

```python
Food # ['Rice', 'Ugali', 'Chapati', 'Pancake']
```

```python
Food.extend(["Matoke", "Nyama Choma"]) # extend our list with another list
Food

# ['Rice', 'Ugali', 'Chapati', 'Pancake', 'Matoke', 'Nyama Choma']
```

```python
print("My fave food is " + Food[4] + ".")  # notice no apostrophes around food[4]

# My fave food is Matoke.
```

```python
len(Food) # count items in food list
# 6
```

```python
Numbers = [15, 40, 50, 100, 115, 140]

print("The fourth element of the Numbers list is" , Numbers[3]) # The fourth element of the Numbers list is 100
```

```python
Food[1:3]   # slice to get the items in the list that you need

# ['Ugali', 'Chapati']
```

```python
Food[:2]  # get first two elements

# ['Rice', 'Ugali']
```

```python
Food[4:]  # get items from index 4 moving on

# ['Matoke', 'Nyama Choma']
```

```python
Food[-2:] # 2nd last item and last

# ['Matoke', 'Nyama Choma']
```

```python
Food.index("Chapati")  # we want to get the index position of chapati

# 2
```

```python
New_Food = ["Anjera", "Chips"]

New_Food  # ['Anjera', 'Chips']
```

```python
Menu = [Food, New_Food]  # joining two lists
Menu

# [['Rice', 'Ugali', 'Chapati', 'Pancake', 'Matoke', 'Nyama Choma'], ['Anjera', 'Chips']]
```

```python
Food.sort() # arrange alphabetically
Food

# ['Chapati', 'Matoke', 'Nyama Choma', 'Pancake', 'Rice', 'Ugali']
```

```python
Food.sort(reverse = True) # arrange alphabetically reversed
Food

# ['Ugali', 'Rice', 'Pancake', 'Nyama Choma', 'Matoke', 'Chapati']
```

```python
Numbers = [1, 3, 4, 5, 2]

Numbers.sort() # ascending order

Numbers

# [1, 2, 3, 4, 5]
```

```python
Numbers.sort(reverse = True) # descending order
```

```python
Numbers

# [5, 4, 3, 2, 1]
```
