```python
import numpy as np
```

```python
m1 = np.array([[5,1,9], [2,8,1]])
m1

#array([[5, 1, 9],
#      [2, 8, 1]])
```

```python
m2 = np.array([[5,1,6], [7,4,1]])
m2

# array([[5, 1, 6],
#      [7, 4, 1]])
```

```python
m3 = np.array([m1, m2])
m3

#array([[[5, 1, 9],
#        [2, 8, 1]],

#       [[5, 1, 6],
#        [7, 4, 1]]])
```

```python
# Data shape

m3.shape

# (2,2,3) => 2 matrices each with 2 rows and 3 columns
```
