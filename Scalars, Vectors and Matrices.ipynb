```python
import numpy as np
```

```python
# scalar

s = 5
```

```python
s 

#5
```

```python
# Vector

v = np.array([5, 3, -1])
```

```python
v

# array([ 5,  3, -1])
```

```python
# matrices

m = np.array([[5, 3, -1], [0, 1,3]])
m

# array([[ 5,  3, -1],
#    [ 0,  1,  3]])
```
