```python
import pandas as pd
```

```python
prices_per_category = {"BlueBand": 100, "Jellywand" : 20, "Bread" : 50}
prices_per_category

# {'BlueBand': 100, 'Jellywand': 20, 'Bread': 50}
```

```python
type(prices_per_category)

# dict
```

```python
prices_per_category = pd.Series(prices_per_category)
prices_per_category

# BlueBand     100
# Jellywand     20
# Bread         50
# dtype: int64
```

```python
type(prices_per_category)

# pandas.core.series.Series
```

```python
prices_per_category.index

# Index(['BlueBand', 'Jellywand', 'Bread'], dtype='object')
```

```python
type(prices_per_category.index)

# pandas.core.indexes.base.Index
```
