```python
def add_ten(x):
    if x >= 100:
        x = x + 10
        return x
    else:
        print("Save more buddy")

add_ten(10) # Save more buddy
```

```python
add_ten(110) # 120
```

```python
def compare_the_two(x, y):
    if (x > y):
        print(str(x)  + " is Greater")
    elif (x < y):
        print(str(x )+  " is Less")
    else:
        print("Equal")

compare_the_two(2, 9) # 2 is Less
```
