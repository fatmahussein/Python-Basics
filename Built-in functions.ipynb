```python
def distance_from_zero(x) :
    if type(x) == int or type(x) == float :          # check positive first
        return  abs(x)
    else: 
        return "Not possible"
        
result1 = distance_from_zero(-3)

result2 = distance_from_zero("cat")

print(result1)

print(result2)
```
